```python
import pandas as pd

df = pd.read_csv("heart.csv")

df.head()
```

```python
import matplotlib.pyplot as plt

df['age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Description:

This histogram shows the distribution of patients' age. Most patients are between 50 and 65 years old.

```python
df['sex'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Description:

This chart shows the number of male and female patients. There are more male patients than female patients.

```python
df['chol'].hist()
plt.title("Cholesterol Distribution")
plt.xlabel("Cholesterol")
plt.ylabel("Count")
plt.show()
```

Description:

This histogram shows the cholesterol levels of patients. Most patients have cholesterol levels between normal and high.

```python
df['target'].value_counts().plot(kind='bar')
plt.title("Heart Disease Distribution")
plt.xlabel("Heart Disease (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()
```

Description:

This chart shows the number of patients with and without heart disease.

```python
df['trestbps'].hist()
plt.title("Resting Blood Pressure Distribution")
plt.xlabel("Resting Blood Pressure")
plt.ylabel("Count")
plt.show()
```

Description:

This histogram shows the distribution of resting blood pressure among patients. Most patients have resting blood pressure between 120 and 140, which is considered normal to slightly high.

## Conclusion

In this assignment, Exploratory Data Analysis (EDA) was performed on the heart disease dataset. The analysis included visualization of age, gender, cholesterol, blood pressure, and heart disease distribution.

The results showed that most patients are between 50 and 65 years old, and there are more male patients than female patients. Cholesterol and blood pressure levels vary among patients.

Overall, the analysis helped in understanding the dataset and identifying important patterns related to heart disease.
